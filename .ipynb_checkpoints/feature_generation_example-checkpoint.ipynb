# Load data

```python
import pandas as pd
import numpy as np

df = pd.read_csv('example_data/example_data.csv')
```

# Time
Create seperate columns for time features

```python
df['created_at'] = pd.to_datetime(df['created_at'])
df['weekday'] = df['created_at'].map(lambda x: x.weekday())
df['hour'] = df['created_at'].map(lambda x: x.hour)
df['date'] = pd.to_datetime(df['created_at']).map(lambda x: str(x).split(' ')[0].replace('-',''))
df['week'] = df['created_at'].map(lambda x: x.week)
df['dayofyear'] = df['created_at'].map(lambda x: x.dayofyear)
df['month'] = df['created_at'].map(lambda x: x.month)

df[['created_at','weekday','hour','week','dayofyear','month','date']].head()
```

For some classifiers features require to be one hot encoded, this can easiliy be done with pandas and sklearn, as can be seen in the code snippit below

https://scikit-learn.org/stable/modules/generated/sklearn.preprocessing.OneHotEncoder.html

```python
one_hot = ['weekday', 'hour', 'month']

for c in one_hot:
    keep = list(df.columns)
    df = pd.concat([df[keep],pd.get_dummies(df[c], prefix='time_'+c)],axis=1)

df[['time_weekday_0', 'time_weekday_1', 'time_weekday_2', 'time_weekday_3', 'time_weekday_4', 'time_weekday_5', 'time_weekday_6', 'time_hour_2', 'time_hour_4', 'time_hour_6', 'time_hour_7', 'time_hour_8', 'time_hour_9', 'time_hour_10', 'time_hour_11', 'time_hour_12', 'time_hour_13', 'time_hour_14', 'time_hour_15', 'time_hour_16', 'time_hour_17', 'time_hour_18', 'time_hour_19', 'time_hour_20', 'time_month_1', 'time_month_2', 'time_month_3', 'time_month_4', 'time_month_8', 'time_month_9', 'time_month_10', 'time_month_11', 'time_month_12']].head()
```

# Weather

Convert hourly datetime to weather recorded at schiphol airport

Data source and information about columns can be found at: https://projects.knmi.nl/klimatologie/uurgegevens/selectie.cgi


```python
weather = pd.read_csv('example_data/weather.csv')
weather.columns = weather.columns.str.replace(' ', '')

def get_weather_dict(date,hour):
    '''
    Input: date and hour
    Output: the weather at the given date and hour
    '''

    weather_at_time = weather[weather['YYYYMMDD'].astype(str)==date][weather['HH'].astype(int)==hour+1].reset_index(drop=True)
    weather_at_time = weather_at_time[['IX', 'M', 'R', 'S', 'O','Y','DD', 'FH', 'FF', 'FX', 'T', 'TD', 'SQ','Q', 'DR', 'RH', 'P', 'U']]
    
    weather_dict = {}
    for key, value in weather_at_time.to_dict().items():
        try:
            weather_dict['weather_'+key] = value[0]
        except:
            weather_dict['weather_'+key] = 0
    return weather_dict

df['weather_dict'] = df.apply(lambda row: get_weather_dict(row.date,row.hour), axis=1)
for column in list(df['weather_dict'][0].keys()):
    df[column] = df['weather_dict'].map(lambda d: d[column])
del df['weather_dict']

df[['weather_IX', 'weather_M', 'weather_R', 'weather_S', 'weather_O', 'weather_Y', 'weather_DD', 'weather_FH', 'weather_FF', 'weather_FX', 'weather_T', 'weather_TD', 'weather_SQ', 'weather_Q', 'weather_DR', 'weather_RH', 'weather_P', 'weather_U']].head()
```

# Geo semantic

Creating a profile of the density and proximity near a coordinates 

The example data used can be found at: https://maps.amsterdam.nl/open_geodata/?LANG=en

```python

def haversine_est(lon1, lat1, lon2, lat2):
    """
    Estimation with equirectangular distance approximation. 
    
    Source: https://www.kaggle.com/justfor/speedup-haversine
    """
    lon1, lat1, lon2, lat2 = map(np.radians, [lon1, lat1, lon2, lat2])

    x = (lon2 - lon1) * np.cos( 0.5*(lat2+lat1) )
    y = lat2 - lat1
    km = 6371.0 * np.sqrt( x*x + y*y )
    return km

def generate_geo_dict(semantic_name,lat,lng,semantic_lats,semantic_lngs):
    '''
    Input:
    A. Latitude and Longitude
    B. Latitudes and longitudes
    
    Output:
    A dictionary that embeds the density and proximity of geo objects in B close to A
    '''
    
    geo_dict = {}

    distances = []
    for i in range(len(semantic_lats)):
        distances.append(haversine_est(lat,lng,semantic_lngs[i], semantic_lats[i]))
    distances = sorted(distances)

    geo_dict['geo_'+semantic_name+'_nearest'] = distances[0]
    geo_dict['geo_'+semantic_name+'_nearest_5_mean'] = np.mean(distances[:5])
    geo_dict['geo_'+semantic_name+'_nearest_10_mean'] = np.mean(distances[:10])
    geo_dict['geo_'+semantic_name+'_nearest_100_mean'] = np.mean(distances[:100])

    geo_dict['geo_'+semantic_name+'_within_25m']  = 0
    geo_dict['geo_'+semantic_name+'_within_50m']  = 0
    geo_dict['geo_'+semantic_name+'_within_100m'] = 0
    geo_dict['geo_'+semantic_name+'_within_200m'] = 0
    
    for distance in distances:

        if distance <0.025:
            geo_dict['geo_'+semantic_name+'_within_25m'] += 1
        if distance <0.05:
            geo_dict['geo_'+semantic_name+'_within_50m'] += 1 
        if distance <0.1:
            geo_dict['geo_'+semantic_name+'_within_100m'] += 1 
        if distance <0.2:
            geo_dict['geo_'+semantic_name+'_within_200m'] += 1    
        
    return geo_dict
    
def load_geo_features(df,column,csv,minimal_number=100,number_of_objects=100000):
    '''
    Creating a geographical fingerprint, by using geo objects, of the area based on the latitude and longitude in the dataframe
    '''
    
    df_geo_objects = pd.read_csv(csv)[:number_of_objects]

    keep = df_geo_objects[column].value_counts().to_frame()
    keep = list(keep[keep[column]>minimal_number].index)
    print('loading', len(keep),'types of ',csv)

    for t in keep:
        lats = list(df_geo_objects[df_geo_objects[column]==t]['lat'])
        lngs = list(df_geo_objects[df_geo_objects[column]==t]['lng'])
        t = t.lower()
        df['geo_'+t+'_dict'] = df.apply(lambda row: generate_geo_dict(t,row.lat,row.lng,lats,lngs), axis=1)
        
        for c in list(df['geo_'+t+'_dict'][0].keys()):
            df[c] = df['geo_'+t+'_dict'].map(lambda d: d[c])
        del df['geo_'+t+'_dict']
    return df

df = load_geo_features(df,'type2','example_data/geo_objects.csv',minimal_number=100,number_of_objects=50000)
```

# Textual features

Create word count features for 5000 words

```python
from nltk.corpus import stopwords
from sklearn.feature_extraction.text import CountVectorizer

def get_top_n_words(corpus, n=None):
    '''
    List the top n words in a vocabulary according to occurrence in a text corpus.
    
    Source: https://gist.github.com/CristhianBoujon/c719ba2287a630a6d3821d37a9608ac8
    '''
    vec = CountVectorizer(stop_words=set(stopwords.words('dutch'))).fit(corpus)
    bag_of_words = vec.transform(corpus)
    sum_words = bag_of_words.sum(axis=0) 
    words_freq = [(word, sum_words[0, idx]) for word, idx in     vec.vocabulary_.items()]
    words_freq = sorted(words_freq, key = lambda x: x[1], reverse=True)
    return words_freq[:n]

words_count = get_top_n_words(df['text'],5000)

df['text'] = df['text'].map(lambda x: x.lower())

for word_count in words_count:
    word = word_count[0]
    count = word_count[1]
    df['text_'+word] = df['text'].map(lambda x: x.count(word)).astype(int)
```

# Visual features

Extracting the 50 visual concepts of each image


```python
from keras import backend as K
from keras.preprocessing import image
from keras.applications.resnet50 import preprocess_input, decode_predictions, ResNet50
from IPython.display import Image

# load model
model = ResNet50(weights='imagenet')


def get_image_output_layer(image_path):
    '''
    Input an image path and output the top 5 visual concepts
    '''

    img = image.load_img(image_path, target_size=(224, 224))
    x = image.img_to_array(img)
    x = np.expand_dims(x, axis=0)
    x = preprocess_input(x)

    preds = model.predict(x)


    return decode_predictions(preds, top=5)[0]

print(get_image_output_layer('images/example3.png'))

Image(filename = 'images/example3.png')
```

# Visual 2024 features
This function can be used to convert any image to a 2024 dimensions representation of the image using imagenet and ResNet

Getting last layer: https://keras.io/getting-started/faq/#how-can-i-obtain-the-output-of-an-intermediate-layer

The hidden representation of the last layer (before the softmax) is often taken as a feature vector, as it contains higher level features

https://github.com/keras-team/keras-applications/blob/master/keras_applications/resnet50.py

```python

def get_layer_before_softmax(image_path):
    '''
    Inputs a file name
    Outputs a 2024 dimensional representation of an image
    '''

    img = image.load_img(image_path, target_size=(224, 224))
    # convert image to numpy array
    x = image.img_to_array(img)
    x = np.expand_dims(x, axis=0)
    x = preprocess_input(x)

    number_of_layers = len(model.layers)
    layer_output = K.function([model.layers[0].input],[model.layers[number_of_layers-2].output])

    return layer_output([x])[0]

print(get_layer_before_softmax('images/example1.jpg'))

Image(filename = 'images/example1.jpg')
```
